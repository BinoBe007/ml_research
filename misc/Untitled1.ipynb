```python
import urllib2
response = urllib2.urlopen('http://haar.thiagohersan.com/haartraining/negatives/')
html = response.read()
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, 'lxml')
```

```python
import os

all_links = soup.find_all("a")
for link in all_links:
    l = link.get("href")
    if '.jpg' in l:
        print('wget http://haar.thiagohersan.com/haartraining/negatives/'+l+' -P /home/wolf/haars/negatives/')
        os.system('wget http://haar.thiagohersan.com/haartraining/negatives/'+l+' -P /home/wolf/haars/negatives/')
```
