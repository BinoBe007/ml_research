```python
from skimage.io import imread
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage.morphology import skeletonize


sample6 = "../step1_image.jpg"
img = rgb2gray(imread(sample6))
plt.subplots(figsize=(20,20))
plt.imshow(img, cmap="gray")
plt.show()
```

```python
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
from skimage.util import invert
from skimage.morphology import skeletonize

def get_binary(img):    
    thresh = threshold_otsu(img)
    binary = img <= thresh
    binary = invert(binary)*1
    return binary

sample_binary = get_binary(img)

plt.subplots(figsize=(20,20))
plt.imshow(sample_binary, cmap="gray")
```

```python
h,w = sample_binary.shape
plt.subplots(figsize=(20,20))
plt.imshow(sample_binary[20:h/3,20:w/3], cmap="gray")
```

```python
template = rgb2gray(imread('../pan1.jpg'))
template_binary = get_binary(template)
h,w = template_binary.shape
plt.subplots(figsize=(20,20))
plt.imshow(template_binary, cmap="gray")
```
