```python
#26th dec 2018
from skimage.io import imread
import matplotlib.pyplot as plt
from skimage.color import rgb2gray

sample6 = "../pan1_scanned_with_data.jpg"
img = rgb2gray(imread(sample6))
plt.subplots(figsize=(20,20))
plt.imshow(img, cmap="gray")
plt.show()
```

```python
#first thing is get the doc and fix the skew.
from skimage.filters import sobel
from skimage.util import invert
import numpy as np
from skimage.transform import rotate

sobel_image = invert(sobel(img))

#find the horizontal projection of all the rows in the image
#rotate the image between angles -10 to 10 degrees and find the angle which gives the least median of horizontal projection
def horizontal_projections(sobel_image):
    sum_of_cols = []
    rows,cols = sobel_image.shape
    for row in range(rows-1):
        sum_of_cols.append(np.sum(sobel_image[row,:]))
    
    return sum_of_cols

rows,cols = sobel_image.shape
predicted_angle = 0
highest_hp = 0
for index,angle in enumerate(range(-10,10)):
    hp = horizontal_projections(rotate(sobel_image, angle, cval=1))
    median_hp = np.median(hp)
    if highest_hp < median_hp:
        predicted_angle = angle
        highest_hp = median_hp

print 'first angle:',predicted_angle
#this first angle gives me an approximate angle where my actual rotation may be present. I will fine tune this 
#further with a better angle.

finetuning_range = np.arange(predicted_angle-1.0, predicted_angle+1.0, 0.01)
for index,angle in enumerate(finetuning_range):
    hp = horizontal_projections(rotate(sobel_image, angle, cval=1))
    median_hp = np.median(hp)
    if highest_hp < median_hp:
        predicted_angle = angle
        highest_hp = median_hp  

print 'fine tuned angle:',predicted_angle

step1_image = rotate(img, predicted_angle, cval=1)

fig, ax = plt.subplots(ncols=2, figsize=(20,10))
ax[0].set_title('original image grayscale')
ax[0].imshow(img, cmap="gray")
ax[1].set_title('original image rotated by angle'+str(predicted_angle))
ax[1].imshow(step1_image, cmap="gray")
```

```python
#now find out how much the image is scaled.
from skimage.io import imread
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
from skimage.util import invert
from skimage.morphology import skeletonize
import numpy as np

def get_binary(img):    
    thresh = threshold_otsu(img)
    binary = img <= thresh
    binary = invert(binary)*1
    return binary

sample_image = step1_image
sample_image = get_binary(rgb2gray(sample_image))

template_image_sample = "../pan1.jpg"
template_img = get_binary(rgb2gray(imread(template_image_sample)))


fig, ax = plt.subplots(figsize=(20,20), nrows=1, ncols=2)
ax[0].imshow(sample_image, cmap="gray")
ax[1].imshow(template_img, cmap="gray")
plt.show()
```

```python
#find the boundaries of the image by using the first pixel found from top, bottom , left and right
def find_boundaries(img):
    r,c = img.shape
    up = []
    lt = []
    rt = []
    bt = []
    #TODO: optimize the below loops.
    #find the upper line
    loops_run = 0
    for ci in range(c):
        found = False
        for ri in range(r):
            loops_run = loops_run + 1
            if not found and img[ri,ci] == 0:
                found = True
                up.append(ri)

    print loops_run

    #find the left line
    for ri in range(r):
        found = False
        for ci in range(c):
            loops_run = loops_run + 1
            if not found and img[ri,ci] == 0:
                found = True
                lt.append(ci)

    print loops_run

    #find the right line
    for ri in range(r):
        found = False
        for ci in xrange(c-1, 0 , -1):#reverse loop
            loops_run = loops_run + 1
            if not found and img[ri,ci] == 0:
                found = True
                rt.append(ci)

    print loops_run

    #find the bottom line
    for ci in range(c):
        found = False
        for ri in xrange(r-1, 0 , -1):#reverse loop
            loops_run = loops_run + 1
            if not found and img[ri,ci] == 0:
                found = True
                bt.append(ri)

    print loops_run
    
    return np.median(lt).astype(int), np.median(rt).astype(int), np.median(up).astype(int), np.median(bt).astype(int)

def crop_image(img, lt, rt, up , bt):
    return img[up: bt, lt:rt]
```

```python
from __future__ import division

#scanned image
lt, rt, up , bt = find_boundaries(sample_image)
sample_image = crop_image(sample_image, lt, rt, up , bt)

#original image
ltt, rtt, upt , btt = find_boundaries(template_img)
template_img = crop_image(template_img, ltt, rtt, upt , btt)

#so how much is it scaled?
scaled_factor = template_img.shape[0]/sample_image.shape[0]
print scaled_factor
```

```python
#need to remove all borders for this guy. i hate these stupid borders!
just_one_image = np.copy(pos_img[:,24:48])

#first let me add some padding to the image so that borders can be captured much more clearly.
def pad_with(vector, pad_width, iaxis, kwargs):
    pad_value = kwargs.get('padder', 10)
    vector[:pad_width[0]] = pad_value
    vector[-pad_width[1]:] = pad_value
    return vector

just_one_image = np.pad(just_one_image, 2, pad_with, padder=1)
```

```python
#let me find all the contour blocks and then figure out the best way to remove borders.
from skimage import measure

# Find contours at a constant value of 0.8
contours = measure.find_contours(just_one_image, 0.8)

print 'contour blocks found: ' , len(contours)

# Display the image and plot all contours found
fig, ax = plt.subplots()
ax.imshow(just_one_image, interpolation='nearest', cmap=plt.cm.gray)
colors = ['green','red','pink','yellow','blue','black']
for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2, c=colors[n])

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

```python
#to see which contour set is closest to the center, i have an idea, 
#lets find the mean distance of all countour points from the center.
from shapely.geometry import Point


r,c = just_one_image.shape
center = Point(r/2,c/2)

least_distant = -1
least_distance = 0
for index, contour in enumerate(contours):
    total_distance = 0
    for cont in contour:
        total_distance += center.distance(Point(cont[0], cont[1]))
        
    mean_distance = total_distance/len(contour)
    
    if least_distant == -1:
        least_distant = index
        least_distance = mean_distance
    elif mean_distance < least_distance:
        least_distant = index
        least_distance = mean_distance
    
print 'least distance one is at index : ' , least_distant

# Display the image and plot all contours found
fig, ax = plt.subplots()
ax.imshow(just_one_image, interpolation='nearest', cmap=plt.cm.gray)
colors = ['black','red','green','yellow','blue','pink']
r,c = just_one_image.shape


ax.plot(contours[least_distant][:, 1], contours[least_distant][:, 0], linewidth=2, c=colors[least_distant])
plt.show()
```

```python
#now that i know which contour is in the center, let me erase the remaining from the canvas.
from shapely.geometry import Point
from shapely.geometry import Polygon
#ref: https://shapely.readthedocs.io/en/stable/manual.html

r,c = just_one_image.shape

for ri in range(r):
    for ci in range(c):
        point = Point((ri,ci))
        polygon = Polygon(contours[least_distant])
        if polygon.contains(point) == False:
            just_one_image[ri][ci] = 1
            
fig, ax = plt.subplots()
ax.imshow(just_one_image, interpolation='nearest', cmap=plt.cm.gray)
```

```python
#now remove the added padding and also resize it 
#center the image in a 28x28 pixel block for Keras Alphabet recognizer.
contour = contours[least_distant].astype(int)
Xmin = np.min(contour[:,0])+1
Xmax = np.max(contour[:,0])+1
Ymin = np.min(contour[:,1])+1
Ymax = np.max(contour[:,1])+1

print Xmin, Xmax, Ymin, Ymax
just_one_image_b = just_one_image[Xmin:Xmax, Ymin:Ymax]
r,c = just_one_image_b.shape
print r,c 
plt.imshow(just_one_image_b, interpolation='nearest', cmap=plt.cm.gray)
plt.show()
```

```python
#center the above image in a 28x28 pixel matrix.
from skimage.transform import rescale
from skimage.util import img_as_float

#first resize the image to 22x22 and then add a padding of 5 pixels around to make a 28x28 pixel
r,c = just_one_image_b.shape
just_one_image_scaled = []
just_one_image_padded = np.ones(28*28).reshape(28,28)
if r > c:
    scale_factor = 20/r
    print 'scaling to ', scale_factor
    just_one_image_scaled = rescale(just_one_image_b.astype(float), scale_factor)
    print just_one_image_scaled.shape
else:
    scale_factor = 20/c
    print 'scaling to ', scale_factor
    just_one_image_scaled = rescale(just_one_image_b.astype(float), scale_factor)
    print just_one_image_scaled.shape
 
#add the scaled image to padded 28*28 matrix
r1, c1 = just_one_image_scaled.shape
r_offset = int((28-r1)/2)
c_offset = int((28-c1)/2)

print r_offset, c_offset

for r in range(r1):
    for c in range(c1):
        just_one_image_padded[r_offset+r, c_offset+c] = just_one_image_scaled[r,c]

just_one_image_padded = get_binary(just_one_image_padded)

#invert the image
r2, c2 = just_one_image_padded.shape
for r in range(r2):
    for c in range(c2):
        if just_one_image_padded[r,c] == 1:
            just_one_image_padded[r,c] = 0
        else:
            just_one_image_padded[r,c] = 1
        

plt.imshow(just_one_image_padded, interpolation='nearest', cmap=plt.cm.gray)
plt.show()
#just_one_image_padded = np.pad(just_one_image_scaled, 2, pad_with, padder=1)
```
