```python
from skimage.io import imread
import matplotlib.pyplot as plt

ori_image = imread('resources/image_with_glass.jpeg')
```

```python
dst = np.array([[0,145],[50,200], [160,75], [100,25]])
fig, ax = plt.subplots()
ax.imshow(ori_image)
ax.scatter(dst[:,0],dst[:,1],color='r')
```

```python
from skimage import transform as tf
import numpy as np

src = np.array([[0, 0], [0, 100], [100, 100], [100, 0]])

tform3 = tf.ProjectiveTransform()
tform3.estimate(src, dst)
warped = tf.warp(ori_image, tform3, output_shape=(100, 100))

plt.imshow(warped, cmap="gray")
```
